```python
# Import the required modules
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
mu = 1
```

```python
#van der pol differential equation
def vdp(state, t, mu):
    x = state[0]
    y = state[1]
    dxdt = y
    dydt = mu*(1-x**2)*y - x
    return [dxdt, dydt]
```

```python
state = [1,2] #intial state
t = np.linspace(0, 100, 100000) #time vector
```

```python
solution = odeint(vdp, state, t, args=(mu,)) #calculate solution

x = solution[:, 0]
y = solution[:, 1]
```

```python
#plot solution as function of time
plt.plot(t, x, t, y)
plt.xlabel('t')
plt.legend(('x', 'y'))
```
